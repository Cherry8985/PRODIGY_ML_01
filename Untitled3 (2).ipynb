```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("train.csv")
print("Shape of the DataFrame:", df.shape)
```

```python
print(df.head(15))
```

```python
print("Information about the DataFrame:")
print(df.info())
```

```python
d2 = df[['LotArea', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'SalePrice']]
print("Selected columns:")
print(d2.head())
```

```python
features = df[['LotArea', 'BedroomAbvGr', 'FullBath', 'HalfBath']]
target = df['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error: {rmse}")
```

```python
plt.scatter(y_test, predictions)
plt.plot([min(y_test), max(predictions)], [min(y_test), max(predictions)], color='green', linestyle='-')
plt.xlabel("Net price")
plt.ylabel("Estimated Prices")
plt.title("Net prices vs Estimated Prices - Linear Regression")
plt.show()
```
